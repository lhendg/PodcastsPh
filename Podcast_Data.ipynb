```python
import pandas as pd
import requests
```

```python
data = pd.read_csv('df_01.csv')
```

```python
ID = data['iTunes URL']
```

```python
a = []
for i in ID:
    a.append(i.split('/')[6])
```

```python
b = []
for x in a:
    b.append(x.split('?')[0])
```

```python
ids = []
for y in b:
    ids.append(y.split('id')[1])
```

```python
ids
```

```python
url = 'https://itunes.apple.com/lookup?id='
```

```python
links = []
for z in ids: links.append(url+str(z))
```

```python
response = requests.get(links[0]) 
data = response.json()
results = data['results']
x = results[0]
```

```python
df = pd.DataFrame(x)
df = df.drop(0)
df = df.drop(1)
```

```python
for link in links:
    response = requests.get(link) 
    data = response.json()
    results = data['results']
    x = results[0]
    df = df.append(x, ignore_index=True)
```

```python
df.to_excel('Podcast_data.xlsx')
```
