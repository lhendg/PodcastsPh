```python
import pandas as pd
import requests
```

```python
data = pd.read_excel('data/Spotify_Top Charts.xlsx')
```

```python
data.shape
```

```python
data = data.dropna()
data.shape
```

```python
data.dtypes
```

```python
ID = data['ID'].astype(int, errors='ignore')
```

```python
url = 'https://itunes.apple.com/lookup?id='
```

```python
links = []
for i in ID: links.append(url+str(i))
```

```python
response = requests.get(links[0]) 
data = response.json()
results = data['results']
x = results[0]
```

```python
df = pd.DataFrame(x)
df = df.drop(0)
df = df.drop(1)
```

```python
for link in links:
    response = requests.get(link) 
    data = response.json()
    results = data['results']
    x = results[0]
    df = df.append(x, ignore_index=True)
```
